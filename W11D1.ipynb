```python
#1.6 quali sono gli anni di inserimento presenti?
singoli_anni = mp["Anno inserimento"].unique()
print(singoli_anni )
```

```python
#1.7 quante attività ci sono nel quadrato di longitudine 9-10 e latitudine 45-46?
long_min= mp["Longitudine"] >= 9
long_max= mp["Longitudine"] < 10
lat_min=mp["Latitudine"] >=45
lat_max=mp["Latitudine"]< 46
int_long = long_min & long_max
int_lat = lat_min & lat_max
sqr = int_long & int_lat


mp["Identificatore in OpenStreetMap"].loc[sqr].count()
```

```python
# in alternativa posso utilizzare il metodo .between()
filtro = mp['Longitudine'].between(9, 10)
filtro2 = mp['Latitudine'].between(45, 46)

mp["Identificatore in OpenStreetMap"].loc[filtro & filtro2].count()
```

```python
#1.8 quante attività ci sono nella provincia di Vicenza?
prov_vi = mp["Provincia"] == "VICENZA"

mp["Provincia"].loc[prov_vi].count()
```

```python
#1.9 quante enoteche ci sono, e come si chiamano? 
lista_enoteche = []
termini = ["enoteca", "vino", "bottiglia", "vini"]
for nome in mp["Nome"]:
    for termine in termini:
        if pd.notna(nome):
            if termine in nome.lower():
                lista_enoteche.append(nome)
                print(nome)

print(len(lista_enoteche))
```

```python
#1.10 quante attività ci sono in Lazio e Abruzzo assieme?
regioni_di_interesse = ["Lazio", "Abruzzo"]
filtro_regioni = mp.loc[:,"Regione"].isin(regioni_di_interesse)

mp["Regione"].loc[filtro_regioni].count()
```

# BOSTON.csv
## Nei beginner_datasets, il file boston.csv contiene il Boston Housing Dataset, che deriva dalle informazioni raccolte dal Census Service degli Stati Uniti sulle abitazioni nell'area di Boston. 

Di seguito vengono descritte le colonne del dataset:

CRIM -tasso di criminalità pro capite per città 

ZN - proporzione di terreni residenziali suddivisi in zone per lotti superiori a 25.000 piedi quadrati. 

INDUS - percentuale di acri di attività commerciali non al dettaglio per città. 

CHAS -variabile dummy del fiume Charles (1 se il tratto costeggia il fiume; 0 altrimenti)

NOX -concentrazione di ossidi di azoto (parti per 10 milioni). 

RM -numero medio di stanze per abitazione 

AGE -proporzione di unità abitative occupate da proprietari costruite prima del 1940 

DIS - distanze ponderate da cinque centri occupazionali di Boston 

RAD -indice di accessibilità alle autostrade radiali 

TAX -aliquota dell'imposta fondiaria sul valore pieno per 10.000 dollari

PTRATIO - rapporto alunni-insegnanti per città 

BLACK -la percentuale di neri per città 

LSTAT - % di popolazione di condizione più bassa 8 

MEDV -Valore mediano delle case, espresso in migliaia di dollari

5.1 La media del prezzo delle case cambia a seconda della distanza dal fiume Charles? 

5.2 Si nota una correlazione tra il tasso di criminalità e il valore delle abitazioni? Come si può spiegare il risultato? 

5.3 Qual è la media del numero di stanze rispetto al rapporto alunni-insegnanti? E del valore delle case? Appare esserci una qualche correlazione? Come si può spiegare il risultato? 

5.4 Rispetto all'accessibilità alle autostrade, cambia qualcosa la media delle età delle abitazioni? E del numero di stanze? E delle tasse?

```python
path = ("../CorsoEpicode/Datasets/boston.csv")
bo = pd.read_csv(path)
bo
```

```python
#5.1 La media del prezzo delle case cambia a seconda della distanza dal fiume Charles?

bo.groupby("chas")["medv"].mean()
```

Direi che il prezzo varia in modo più o meno significativo, poiché la vicinanza al fiume comporta un prezzo superiore al 30& ca. rispetto alle case che non lo costeggiano.

```python
#5.2 Si nota una correlazione tra il tasso di criminalità e il valore delle abitazioni? Come si può spiegare il risultato?

bo.corr(numeric_only = True)
```

Abbiamo una correlazione negativa, più o meno importante. La presenza di criminalità abbassa il valore degli immobili in modo sufficientemente significativo. 

```python
#5.3 Qual è la media del numero di stanze rispetto al rapporto alunni-insegnanti? 
#E del valore delle case? 
#Appare esserci una qualche correlazione? 
#Come si può spiegare il risultato?

print(bo.groupby("rm")["ptratio"].mean())

print(bo.groupby("rm")["medv"].mean())

bo[["rm","medv","ptratio","lstat"]].corr()

#R: Facendo una piccola ricerca: "In generale, esiste una correlazione positiva tra il costo medio delle case e il rapporto alunni/insegnanti, 
#ma non è una relazione diretta e semplice. Un basso rapporto alunni/insegnanti, 
#indice di classi meno affollate e maggiore attenzione individuale, 
#può essere visto come un fattore positivo per la qualità dell'istruzione, 
#e questo può portare ad un aumento della domanda di case in quella specifica area, 
#con conseguente aumento dei prezzi immobiliari. 
#Tuttavia, il costo delle case è influenzato da molti altri fattori, 
#tra cui la vicinanza a servizi, la qualità della vita, la reputazione delle scuole e il mercato immobiliare generale" 
```

```python
#5.4 Rispetto all'accessibilità alle autostrade, cambia qualcosa la media delle età delle abitazioni? E del numero di stanze? E delle tasse?

print(round(bo.groupby("rad")["age"].mean(), 2))

for r in bo["rad"].unique():
  print(r, round(bo[bo["rad"]==r]["age"].mean(), 2))

bo[["rad","tax","rm","age"]].corr()
```

Le tabelle dimemployee e dimemployeesalesterritory come DataFrame

```python
import os
import dotenv
import sqlalchemy
import pandas as pd

dotenv.load_dotenv(dotenv_path="cred.env", override=True)

username = os.getenv("username")
password= os.getenv("password")
host=os.getenv("host")
dbname=os.getenv("dbname")

connection_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"
print(connection_string)

db_engine = sqlalchemy.create_engine (connection_string) 

query = "SELECT * FROM dimemployee"
dimemployee = pd.read_sql(query, db_engine)

dm = dimemployee

query2 = "SELECT * FROM dimemployeesalesterritory"
dimemployeesalesterritory = pd.read_sql(query2, db_engine)

ds = dimemployeesalesterritory
ds.info()
```

```python
#7.1 Effettuiamo un join tra i due DataFrame usando le colonne EmployeeKey
dm1 = dm.set_index("EmployeeKey")
ds1 = ds.set_index("EmployeeKey")
joined = dm1.join(ds1)
joined
```

```python
#7.2 Controlliamo la dimensione del DataFrame risultante: è quella attesa?
joined.shape
#R: Me lo aspettavo più grande.
```

```python
#7.3 Importiamo ora la tabella dimsalesterritory ed effettuiamo un join tra questa e il DataFrame risultante della join precedente, usando le colonne SalesTerritoryKey

query3 = "SELECT * FROM dimsalesterritory"
dimsalesterritory = pd.read_sql(query3, db_engine)
dt = dimsalesterritory
dt

newjoin=joined.merge(dt, on="SalesTerritoryKey")
newjoin
```

```python
#7.4 Su questo DataFrame contare quanti dipendenti ci sono per ogni paese (country) e per ogni regione (region)

print(newjoin.groupby("SalesTerritoryCountry")["EmailAddress"].count())
newjoin.groupby("SalesTerritoryRegion")["EmailAddress"].count()
```

```python
#7.5 Valutiamo la media del BaseRate per ogni paese: ci sono differenze?

newjoin.groupby("SalesTerritoryRegion")["BaseRate"].count()
```
