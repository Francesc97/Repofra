```python
#1.9 Quanto pesa il più pesante? 
dimproduct["Weight"].max()
```

```python
#1.10 Quanti prodotti pesano più di 100? 
filtro_peso = dimproduct["Weight"] > 100
dimproduct.loc[filtro_peso]      #aggiungendo .shape ottengo il numero delle rows presenti con il masking
```

```python
#1.11 Quanto costano in media i prodotti (colonna DealerPrice)? 
dimproduct["DealerPrice"].mean()
```

```python
#1.12 Quali sono i quartili dei prezzi?
print(dimproduct["DealerPrice"].quantile(0.25))
print(dimproduct["DealerPrice"].quantile(0.50))
print(dimproduct["DealerPrice"].quantile(0.75))
print(dimproduct["DealerPrice"].quantile(1.00))
```

```python
#1.13 Qual è il prezzo medio per i prodotti di colore blu? 
filtro_colore = dimproduct["Color"] == "Blue"
dimproduct["DealerPrice"].loc[filtro_colore].mean()
```

```python
#1.14 Qual è il prezzo medio per i prodotti di colore rosso o nero? 
colore1 = dimproduct["Color"] == "Red"
colore2 = dimproduct["Color"] == "Black"
dimproduct["DealerPrice"].loc[colore1|colore2].mean()
```

```python
#1.15 Qual è il prezzo massimo per i prodotti di taglia 42 e peso oltre i 10 Kg? 
filtro_taglia = dimproduct["Size"] == "42"
filtro_peso = dimproduct["Weight"] > 10
dimproduct["DealerPrice"].loc[filtro_taglia & filtro_peso].max()
```

```python
#1.16 Qual è il nome inglese e il costo di produzione (StandardCost) di tutti i prodotti di taglia 42, peso oltre i 10 Kg e colore argento? 
colore3 = dimproduct["Color"] == "Silver"
dimproduct.loc[::,["EnglishProductName","StandardCost"]].loc[filtro_taglia & filtro_peso & colore3]
```

```python
#1.17 Visualizziamo lo StandardCost e il DealerPrice degli ultimi 20 elementi del dataset: quali sono le differenze? C'è un pattern? C'è qualche elemento che non lo segue?

dimproduct.tail(20)[["StandardCost","DealerPrice"]]

#R: Sembra esserci una proporzionalità diretta tra lo standardcost e il dealerprice, eccetto per le ultime 3 righe, in cui il prezzo di vendita è inferiore a quello di produzione 
#(ipotesi: prodotto in svendita per previsione errata di "vendite pregresse"?)
```
