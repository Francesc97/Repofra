# Dal database AdventureWorks 

estraiamo la tabella dimproduct 

1.1 Sulla colonna DealerPrice, utilizzando il metodo .round(), arrotondiamo i valori alle due cifre decimali, e poi al valore intero più vicino 

1.2 Utilizzando il metodo .clip(), facciamo in modo che i valori siano compresi tra un minimo di 0 e un massimo di 1000

```python
import os
import dotenv
import sqlalchemy
import pandas as pd

dotenv.load_dotenv(dotenv_path="cred.env", override=True)

username = os.getenv("username")
password= os.getenv("password")
host=os.getenv("host")
dbname=os.getenv("dbname")

connection_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"
print(connection_string)

db_engine = sqlalchemy.create_engine (connection_string) 

query = "SELECT * FROM dimproduct"
dimproduct = pd.read_sql(query, db_engine)
```

```python
#1.1 Sulla colonna DealerPrice, utilizzando il metodo .round(), 
#arrotondiamo i valori alle due cifre decimali, e poi al valore intero più vicino

dp = dimproduct

dp_round2 = dp["DealerPrice"].round(2) 
dp_round = dp["DealerPrice"].round() 
dp_round
```

```python
#1.2 Utilizzando il metodo .clip(), facciamo in modo che i valori siano compresi tra un minimo di 0 e un massimo di 1000

dp["DealerPriceClipped"] = dp["DealerPrice"].clip(lower = 0, upper = 1000)

dp.sample(5)
```

```python
import pandas as pd
```

```python
path = ("../CorsoEpicode/Datasets/facebook.csv")
fb = pd.read_csv(path)
fb
```

```python
#4.1 Con la funzione pd.to_datetime() convertiamo la colonna status_published in formato Timestamp
import pandas as pd

fb["status_published"] = pd.to_datetime(fb["status_published"])
fb["status_published"].info()
```

```python
#4.2 Utilizzando gli attributi .dt.year , .dt.month , .dt.day , .dt.dayofweek , .dt.dayofyear,
#ottieniamo informazioni specifiche sulle date delle transazioni, come l'anno, il mese, il giorno della settimana, il giorno dell'anno, eccetera
fb.loc[:, "Month"] = fb.loc[:,"status_published"].dt.month
fb

fb.loc[:, "Year"] = fb.loc[:,"status_published"].dt.year

fb.loc[:, "Day"] = fb.loc[:,"status_published"].dt.day

fb.loc[:, "WeekDay"] = fb.loc[:,"status_published"].dt.dayofweek

fb.loc[:, "DayOfYear"] = fb.loc[:,"status_published"].dt.dayofyear

fb
```

```python
#4.3 Estraiamo solo i post relativi al 2012

anno = fb["Year"] == 2012

fb.loc[anno].describe()
```

```python
#4.4 Estraiamo solo i post relativi a maggio 2018
anno_2018 = fb["Year"] == 2018
mese_maggio = fb["Month"] == 5

fb.loc[anno_2018 & mese_maggio]
```

```python
#4.5 Confrontiamo il numero di post pubblicati nei weekend rispetto al numero di post pubblicati nel resto della settimana

week = fb["WeekDay"].between(0, 4)
weekend = fb["WeekDay"].between(5, 6)

print(fb.loc[week].count()) 
fb.loc[weekend].count()
```

```python
#4.6 Troviamo il primo e ultimo post pubblicati in ogni anno
fb_sorted = fb.sort_values('status_published')
display(fb_sorted.groupby("Year")[["status_published", "status_id"]].agg(["first", "last"]))

display(fb.groupby("Year").first())
fb.groupby("Year").last()
```

```python
#ALTERNATIVA
fb_sorted = fb.sort_values('status_published')
grouped = fb_sorted.groupby('Year')

# Combina primo e ultimo
primo = grouped.first().reset_index()
primo['posizione'] = 'primo'

ultimo = grouped.last().reset_index()
ultimo['posizione'] = 'ultimo'

risultato = pd.concat([primo, ultimo]).sort_values('Year')
risultato
```

```python
#4.7 Quanti tipi di post ci sono? E quanti per ogni tipo?
print(f" Ci sono {fb["status_type"].nunique()} tipi di post:")

fb.groupby("status_type")["status_type"].count()
```

Nel pacchetto os della standard library c'è la funzione os.listdir() che permette di avere la lista dei nomi di file all'interno di una directory; senza input di default li cerca nella directory di lavoro corrente, altrimenti si può passare un path per esaminare una directory specifica, ad esempio os.listdir("mio_progetto/beginner_datasets/") 

 Nella directory dei beginner_datasets, quali sono i dataset che contengono dati nulli?

8.1 Dovremo usare un ciclo for per esaminare tutti i nomi dei file 

8.2 Dovremo selezionare solo i nomi di file con estensione .csv (quindi usare un costrutto if) 

8.3 Nel corpo dovremo leggere di volta in volta il file in esame, e caricarlo in un DataFrame con la funzione .read_csv() 

8.4 Sul DataFrame dovremo utilizzare il metodo .isna() per trovare la maschera booleana dei dati nulli 

8.5 Dovremo contare i dati nulli, utilizzando .sum(); potremmo doverlo utilizzare più di una volta 

8.6 Dovremo stampare, o memorizzare in una list, solo i nomi dei file che contengono dati nulli

```python
import os 
files_with_nulls = []

for filename in os.listdir("../CorsoEpicode/Datasets"):
    if filename.endswith(".csv"):
        df = pd.read_csv(f"../CorsoEpicode/Datasets/{filename}")
        if df.isna().sum().sum() > 0:      #doppio sum() perhcé sommoo per colonne (prima) e sommo nuovamente per ottenere i nulli totali nel df
            files_with_nulls.append(filename)
            print(f"{filename} ha dati nulli")

print(f"File con dati nulli: {files_with_nulls}")

os.listdir("../CorsoEpicode/Datasets")
```
