Sul database titanic, effettuiamo le seguenti analisi: 

1.1 Riusciamo a calcolare quanti ponti c'erano sulla nave (prima lettera dei dati nella colonna Cabin)?

1.2 Visualizzare un grafico con il numero di passeggeri di ogni classe di imbarco (colonna PClass) 

1.3 Fare la stessa cosa per la colonna Survived 

1.4 Qual era la distribuzione delle tariffe (Fare)? 

1.5 Riusciamo a vedere la distribuzione delle età dei passeggeri rispetto alla classe di imbarco con un boxplot? 

1.6 Visualizziamo un boxplot rispetto alle colonne Fare e Survived; che cose ne deduciamo?

```python
import os
import dotenv
import sqlalchemy
import pandas as pd

dotenv.load_dotenv(dotenv_path="cred.env", override=True)

username = os.getenv("username")
password= os.getenv("password")
host=os.getenv("host")
dbname=os.getenv("dbnametitanic")

connection_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"
print(connection_string)

db_engine = sqlalchemy.create_engine (connection_string) 

query = "SELECT * FROM titanic"
ti = pd.read_sql(query, db_engine)
ti
```

```python
#1.1 Riusciamo a calcolare quanti ponti c'erano sulla nave (prima lettera dei dati nella colonna Cabin)?

ti["Ponte"]= ti["Cabin"].str.extract(r"([A-Z])", expand = False)
ti["Ponte"].value_counts()
```

```python
#1.2 Visualizzare un grafico con il numero di passeggeri di ogni classe di imbarco (colonna PClass)
print(ti.groupby("PClass")["PassengerId"].count())
ti.groupby("PClass")["PassengerId"].count().plot(kind="bar")
```

```python
#1.3 Fare la stessa cosa per la colonna Survived
print(ti.groupby("Survived")["PassengerId"].count())
ti.groupby("Survived")["PassengerId"].count().plot(kind="bar")
```

```python
#1.4 Qual era la distribuzione delle tariffe (Fare)?
ti["Fare"].plot(kind="hist")
```

```python
#1.5 Riusciamo a vedere la distribuzione delle età dei passeggeri rispetto alla classe di imbarco con un boxplot?
import seaborn as sns
sns.boxplot(data=ti, x="PClass", y="Age")
```

```python
#1.6 Visualizziamo un boxplot rispetto alle colonne Fare e Survived; che cose ne deduciamo?

sns.boxplot(data=ti, y="Fare", x="Survived")
```

Scarichiamo il dataset metal_bands_2017.csv dalla piattaforma e: 

3.1 Con Seaborn visualizziamo l'istogramma del numero dei fan delle varie band; cosa si vede? Cose ne deduciamo? 

3.2 Delle band che suonano sia Progressive che Thrash, realizziamo un barplot che mostri quanti fan ci sono in totale rispetto al paese di origine (nota: in questo caso specifico, per accedere alla colonna "style" dovremo usare .loc[:, "style"] e non .style perché quest'ultimo è un metodo) 

3.3 Visualizziamo un istogramma delle date di formazione delle band (potremo dover convertire in qualche modo questo dato); si nota qualche trend?

```python
import os
import numpy as np
import pandas as pd 
```

```python
path = ("../CorsoEpicode/Datasets/metal_bands_2017.csv")
mb = pd.read_csv(path, encoding = "latin1", index_col=0)
mb
```

```python
#3.1 Con Seaborn visualizziamo l'istogramma del numero dei fan delle varie band; cosa si vede? Cose ne deduciamo?
import seaborn as sns

sns.countplot(data = mb, x="fans")
mb["fans"].plot(kind="hist", bins=50)
```

Notiamo che la popolarità delle band è significativa per un numero ristretto delle band all'interno del DataFrame.

```python
#3.2 Delle band che suonano sia Progressive che Thrash, realizziamo un barplot che mostri 
#quanti fan ci sono in totale rispetto al paese di origine 
#(nota: in questo caso specifico, per accedere alla colonna "style" dovremo usare .loc[:, "style"] e non .style perché quest'ultimo è un metodo)

progressive = mb["style"].str.contains("progressive", case = False)
thrash = mb["style"].str.contains("thrash", case = False)

mbfil = mb.loc[progressive & thrash]

display(mbfil.groupby("origin")["fans"].size())

mbfil.groupby("origin")["fans"].size().plot(kind="barh")
```

```python
#3.3 Visualizziamo un istogramma delle date di formazione delle band (potremo dover convertire in qualche modo questo dato); si nota qualche trend?

mb["formed"].value_counts()

filtro = mb["formed"] != "-"

mbfiltrato = mb.loc[filtro]

mbfil["formed"] = mbfiltrato["formed"].astype(int)

mbfil["formed"].plot(kind="hist")
```

Si nota come negli anni le band metal siano "nate" in maniera progressiva (costi inferiori e quindi accessibitlità maggiori agli strumenti, sale prove, ecc) e come questo trend sia calato intorno al 2010, probabilmente in corrispondenza della crescente popolarità di altri generi musicali.
